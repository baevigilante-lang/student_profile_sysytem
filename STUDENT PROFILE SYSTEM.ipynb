```python
full_name= input("Enter your full name: ")
age= int(input("Enter your age: "))
gpa= float(input("Enter your GPA: "))
courses_input= (input("Enter courses enrolled: "))
courses= courses_input.split(",")
courses= [course.strip()for course in courses]
```

```python
student_profile= {
    "name": full_name,
    "age": age,
    "gpa": gpa,
    "courses_input": courses_input,
    "num_courses": len(courses)
}
```

```python
print("STUDENT PROFILE")
print(f"Full Name: {student_profile['name']}")
print(f"Age: {student_profile['age']}")
print(f"GPA:{student_profile['gpa']}")
print(f"Courses_input: {student_profile['courses_input']}")
print (f"Number of courses:{student_profile['num_courses']}")
```
